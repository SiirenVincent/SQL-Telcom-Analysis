**Importing the Libraries to be used**

```
#Import the pandas library
#
import pandas as pd
#Import the Numpy library
import numpy as np
```

**Load Datasets**

```
# Load Telcom_dataset.csv data

tel_1=pd.read_csv('Telcom_dataset.csv')
tel_1.head(5)
```

```
# Load Telcom_dataset2.csv data

tel_2=pd.read_csv('Telcom_dataset2.csv')
tel_2.head(5)
```

```
# Load Telcom_dataset3.csv data

tel_3=pd.read_csv('Telcom_dataset3.csv')
tel_3.head(5)
```

```
# Load cells_geo.csv data

cells_geo=pd.read_csv('cells_geo.csv', delimiter=';')
cells_geo.head(5)
```

### Telcom Datasets Cleaning

```
# correct the colunm produtc in Telcom_dataset.csv to read same as others
tel_1new=tel_1.rename(columns={'PRODUTC':'PRODUCT','DATETIME':'DATE_TIME'}, inplace=False)
tel_1new
```

```
# correct the colunm DW_A_NUMBER to DW_A_NUMBER_INT,,DW_B_NUMBER to DW_B_NUMBER_INT in Telcom_dataset.csv to read same as others
tel_2new=tel_2.rename(columns={'DW_A_NUMBER':'DW_A_NUMBER_INT','DW_B_NUMBER':'DW_B_NUMBER_INT'}, inplace=False)
tel_2new
```

```
# correct the colunm CELLID, to CELL_ID,SIET_ID to SITE_ID in Telcom_dataset.csv to read same as others
tel_3new=tel_3.rename(columns={'CELLID':'CELL_ID','SIET_ID':'SITE_ID'}, inplace=False)
tel_3new
```

**Merge the three datesets of Telcom_datasets**

```
# Merge Telcom_dataset,telcom_dataset2,Telcom_dataset3
tel_123=pd.concat([tel_1new,tel_2new,tel_3new],ignore_index=True)            
tel_123
```

```
# Delete duplicate values
tel_123.duplicated()
tel_dup=tel_123.drop_duplicates()
tel_dup
```

```
# Delete Null data
tel_null=tel_dup.dropna()
tel_null
```

```
# delect column cell_on_site,country_A,country_B in telcome_datasets
dftel=tel_null.drop(['COUNTRY_A','COUNTRY_B','CELL_ON_SITE',
              'DW_A_NUMBER_INT','DW_B_NUMBER_INT','SITE_ID'],axis=1)
dftel
```

**Cell Geo Dataset Cleaning**

```
# Delete duplicate values
cells_geo.duplicated()
cells_geo1=cells_geo.drop_duplicates()
cells_geo1
```

```
# Delete Null values
cells_geo1.isnull()
cells_geo2=cells_geo1.dropna()
cells_geo2
```

```
# delete unnecesary columns
cells_geo2=cells_geo2.drop(['Unnamed: 0','AREA','LONGITUDE','LATITUDE','DECOUPZONE',
                            'ZONENAME','SITE_CODE','STATUS'],axis=1)
cells_geo2
```

### **Merge two datasets cells_geo2 and tell_123**

```
# Merge dftel and cells_geo2 datasets
tel_cells_geo=pd.concat([dftel,cells_geo2],axis=0,ignore_index=True)            
tel_cells_geo
```

```
# Get the most top city that use the infrastruter using 
top_cities=tel_cells_geo['VILLES'].value_counts()
top_cities.head(10)
```

```
# most used between voice,data,sms
tel_cells_geo.groupby(['PRODUCT'])[['VALUE']].sum()
```

```
# find cities that are active day time
time=tel_cells_geo.groupby(['DATE_TIME'])[['VALUE']].sum()
time
time1=time.sort_values(by='VALUE', ascending=0)
time1.head(10)
```
